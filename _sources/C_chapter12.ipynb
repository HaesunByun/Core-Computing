- ì•„ë˜ì˜ ì½”ë“œì—ì„œ key=lenìœ¼ë¡œ ì£¼ë©´ len()í•¨ìˆ˜ê°€ ì „ë‹¬ë˜ì–´ ë‹¨ì–´ì˜ ê¸¸ì´ìˆœìœ¼ë¡œ ì •ë ¬í•˜ë¼ëŠ” ì˜ë¯¸ì´ë‹¤. ì´ë•Œ ê´„í˜¸ì—†ì´ í•¨ìˆ˜ëª…ë§Œ ì ì–´ì¤€ë‹¤.

- ê¸°ì¤€ìœ¼ë¡œ ì‚¬ìš©í•  ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ì„œ ì ìš©í•  ìˆ˜ë„ ìˆë‹¤.
- ì•„ë˜ì˜ ì½”ë“œì—ì„œëŠ” íŠœí”Œì˜ ì¸ë±ìŠ¤ 1ì˜ ê°’ì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ takeSecond()ë¥¼ ì •ì˜í•˜ê³ , ë°˜í™˜ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ ì •ë ¬í•˜ì˜€ë‹¤.

````python
random = [(2, 2), (3, 4), (4, 1), (1, 3)]

def takeSecond(elem):
    return elem[1]

random.sort(key=takeSecond)

print('Sorted list:', random)
````

- ë”•ì…”ë„ˆë¦¬ì˜ ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ ì‚¼ì•„ ì •ë ¬í•  ìˆ˜ë„ ìˆë‹¤.
- get_name()í•¨ìˆ˜ëŠ” ë”•ì…”ë„ˆë¦¬ì˜ í‚¤ì™€ ë§¤í•‘í•˜ëŠ” valueë¥¼ ë°˜í™˜í•œë‹¤. 

````python
employees = [
  {'Name': 'Alan Turing', 'age': 25, 'salary': 10000},
  {'Name': 'Sharon Lin', 'age': 30, 'salary': 8000},
  {'Name': 'John Hopkins', 'age': 18, 'salary': 1000},
  {'Name': 'Mikhail Tal', 'age': 40, 'salary': 15000},
]
````

````python
employees[0].get('Name')
````

- employee.get('Name')ì€ ì´ë¦„ì„ ë°˜í™˜í•œë‹¤. ì¦‰ ì´ë¦„ì„ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œë‹¤. 

````python
def get_name(employee):
    return employee.get('Name')

employees.sort(key=get_name)
print(employees)
````

- get_name()ê³¼ ê°™ì€ í•¨ìˆ˜ë¥¼ ì •ì˜í•˜ëŠ” ëŒ€ì‹  lambda í‚¤ì›Œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ìµëª…í•¨ìˆ˜ë¥¼ ì •ì˜í•˜ì—¬ ì‚¬ìš©í•  ìˆ˜ ìˆë‹¤.

````python
employees.sort(key=lambda x: x.get('Name'))
print(employees)
````
