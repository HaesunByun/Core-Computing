## 리스트 생성
- 리스트는 숫자, 문자, 리스트, 딕셔너리 등 어떠한 데이터 타입이든 다 담을 수 있는 만물박스이다.
- 리스트는 대괄호[   ]로 묶어준다.
- 리스트에 들어있는 데이터를 아이템, 원소, 요소이라고 부른다.
- 리스트 내 아이템은 콤마(,)로 구분한다.

### 직접 생성
- 대괄호 안에 아이템을 콤마로 나열하여 직접 생성할 수 있다.

````python
a = [1, 2, 3, 4, 5]
b = [3, 8, 1, 3, 2]
c = [1, 'apple', ['banana',1, 2], 'coconut']
d = ['apple', 'banana', 'coconut', 'durian']
e = [ ] 
````

### list() 함수로 생성
- list() 함수는 빈 리스트, 숫자리스트, 문자열을 낱개로 분리하여 리스트로 만드는 함수이다. 

````python
item = list()
````

````python
a=list(range(1,10))
print(a)
````

### split() 메서드로 생성
- split() 함수는 문자열을 특정 기준에 따라 분리하여 리스트로 만드는 함수이다. 형식은 다음과 같다.

```
문자열.split(분리기준 문자열)
```
- 문자열을 괄호안의 '분리기준 문자열'을 기준으로 분리하며, 괄호안에 아무것도 없으면 공백을 기준으로 분리한다.

````python
item = '바나나 딸기 복숭아'
fruit = item.split()
print(fruit)
````

````python
item = '바나나, 딸기, 복숭아'
fruit = item.split(', ')
print(fruit)
````

## 인덱싱(Indexing)

- **인덱스**는 리스트 내 아이템의 주소이다. 즉, 아이템에 접근하기 위해 인덱스를 사용한다.

### 양의 인덱스(positive index)
positive index는 가장 왼쪽 아이템에서 0부터 시작하여 오른쪽으로 1씩 증가한다.

````python
shape= ['turtle', 'arrow', 'circle', 'square']
print(shape[0])
print(shape[1])
print(shape[2])
print(shape[3])
````

- 리스트안에 리스트가 있는 경우 대괄호를 연속하여 쓰면 된다.
- 예를 들어, 'banana'를 꺼내려면 c[2][0]과 같이 접근할 수 있다.

````python
c=[1, 'apple', ['banana',1, 2], 'coconut']
print(c[2])
print(c[2][0])
````

### 음의 인덱스(negative index)
- negative index는 가장 오른쪽 끝 아이템에서 -1부터 시작하여 왼쪽으로 1씩 감소한다.

````python
d=['apple', 'banana', 'coconut', 'durian']
print(d[-1], d[-2], d[-3], d[-4])
````

- 없는 인덱스에 접근하면 에러가 난다.


```python
print(d[4])
print(d[-5])
```


### 아이템 수정
- 인덱스를 이용하여 아이템을 수정할 수 있다.

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
shape[2]=100
print(shape)
````

## 슬라이싱(Slicing)
슬라이싱이란 리스트의 아이템 일부를 자르는 것이다. 형식은 다음과 같다.

```
리스트변수[start:end:step]
```
- start: 슬라이싱 시작 인덱스
- end: 슬라이싱하려는 끝 인덱스+1
- step: 이동간격, 양수이면 오른쪽으로 step만큼 이동한다. 음수이면 왼쪽으로 step만큼 이동한다.

- start, end, step 모두 생략 가능하다.

- 아래의 코드를 통해 실습해보자. 처음부터 인덱스 3 이전까지 오른쪽으로 이동하여 슬라이싱한다.

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
print(shape[ :3])
````

- 인덱스 -3부터 끝까지 오른쪽으로 이동하여 슬라이싱한다.

````python
print(shape[-3: ]) 
````

- 처음부터 인덱스 -3 이전까지 오른쪽으로 이동하여 슬라이싱한다.

````python
print(shape[:-3])
````

- 처음부터 인덱스 -3 이전까지 왼쪽으로 이동하여 슬라이싱한다.

````python
print(shape[:-3:-1])
````

- 처음부터 인덱스 -3 이전까지 2씩 건너뛰면서 왼쪽으로 이동하여 슬라이싱한다.

````python
print(shape[:-3:-2])
````

a = "Life is too short, You need Python"

- 리스트뿐만 아니라 문자열, 딕셔너리, 튜플, 집합 등 len() 함수를 통해 아이템의 개수를 알 수 있다.

## 리스트 주요 메서드

|사용 방법|의미|예시(a=['a','b','c']일 때)|
|:----------:|:----------|:----------|
|`list.append(item)`|list의 마지막에 item을 추가| a.append('d') ➡ ['a','b','c','d'] |
|`list.extend(list2)`|list의 마지막에 list2의 아이템을 추가하여 확장| a.extend(['d','e']) ➡ ['a','b','c','d','e']| 
|`list.insert(index, item)`|list의 index 위치에 item을 삽입| a.insert(1,'d') ➡ ['a','d','c','d']| 
|`list.remove(item)`|list에서 첫 번째로 나오는 item을 삭제|a.remove('c') ➡ ['a','c']|
|`list.index(item)`|list에서 item의 index를 반환|a.index('c') ➡ 2|
|`list.count(item)`|list에서 item의 개수를 반환|a.count('c') ➡ 1|
|`list.clear()`|list의 모든 item의 제거한다.|a.clear() ➡ []|
|`list.pop(index)`|list에서 인덱스의 아이템을 제거하고 반환.<br> 인덱스를 지정하지 않으면 마지막 아이템을 제거하고 반환|a.pop(1) ➡ ['a','c']|
|`list.reverse()`|list에서 아이템의 순서를 뒤집는다.|a.reverse() ➡ ['c','b','a']|
|`list.copy()`|list를 shallow copy한다.|b=a.copy() ➡ b= ['a','b','c']|
|`list.sort()`|list 내 item를 오름차순으로 정렬| a = ['c','a','b']일 때, a.sort() ➡ ['a', 'b', 'c']|
|`list2 = sorted(list)`|list를 오름차순으로 정렬하여 반환|b = sorted(a) ➡  b = ['a', 'b', 'c']|

### 아이템 추가 append(item)

## List comprehension(리스트 축약법)
- 반복문과 조건문을 한 줄의 코드로 축약하여 리스트를 만들 수 있다. 
- 리스트뿐만 아니라 딕셔너리, 집합에서도 축약하여 만들 수 있다.
- 이러한 축약법은 파이썬이 `파이썬스럽다` 라는 것을 보여주는 대표적인 예이다.


````python
print('풀어쓴 코드======')
x=[ ]
for i  in  range(1,11): 
    x.append(i**2)
print(x)

print('축약형 코드======')
x=[i**2 for i  in  range(1,11)]
print(x)
````

````python
print('풀어쓴 코드======')
x=[ ]
for i  in  range(1,11):
     x.append(i/10)
print(x)

print('축약형 코드======')
print([i/10 for i in range(1,11)])
````

````python
print('풀어쓴 코드======')
x=[ ]
for i  in  ['70','90','100']:
     x.append(int(i))
print(x)

print('축약형 코드======')
print([int(i) for i in ['70','90','100']])
````
