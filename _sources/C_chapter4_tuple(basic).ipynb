## set ì£¼ìš” ë©”ì„œë“œì™€ iteration

|ì‚¬ìš© ë°©ë²•|ì˜ë¯¸|ì˜ˆì‹œ<br> s1={1,4,7} b={1,3,6,8,10}|
|:----------:|:----------|:----------|
