### Problem: 가까운 이웃 그래프로 나타내기

😄  윗 문제에서 찾은 가까운 이웃을 그래프로 나타내보세요. 아래와 같은 matplotlib 함수들을 사용할 수 있습니다.
```
plt.scatter(x,y)  : 좌표에 동그란 점을 찍는데 사용합니다.
plt.text('snu') : 그래프에 이름을 출력하는데 사용합니다.
plt.plot(x,y) : 동그란 점사이에 선을 잇는데 사용합니다.
plt.show( ) : 그래프를 나타냅니다.
```


````python
import ast
import matplotlib.pyplot as plt

nameList = ast.literal_eval(input())
locList = ast.literal_eval(input())

resident = []

for name, loc in zip(nameList, locList):
    resident.append((name, loc))
    plt.scatter(*loc)
    plt.text(*loc, name)

for curr in resident:
    # 이웃 탐색을 한줄의 코드로 줄이면
    neighbor = min( list(filter(lambda mm: mm[0] != curr[0], resident)), key = lambda nn: (nn[1][0] - curr[1][0])**2 + (nn[1][1] - curr[1][1])**2)
    plt.plot((curr[1][0], neighbor[1][0]), (curr[1][1], neighbor[1][1]))

plt.show()
````
