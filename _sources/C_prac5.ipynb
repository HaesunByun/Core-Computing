### Problem: 다항식의 곱셈
😄 여러 개의 다항식을 곱해 봅시다.
다항식은 각 항의 계수로 나타낼 수 있습니다. 
예를 들어 '2'는 다항식 2를, '1 3 4'는 다항식 $ 1+3x+4x^2$를, '0 4 0 1'은 다항식 $ 4x+x^3 $을 뜻합니다.
입력으로 주어지는 각 다항식의 계수는 모두 정수이며, 최고차항의 계수는 0이 아닙니다. 
즉 '0'이나 '2 3 0 0', '5 0' 등이 입력으로 들어오지 않습니다. 

입력으로는 다항식의 개수 N 이 주어진 후, N 개의 다항식의 계수가 차례대로 주어집니다.
예를 들어, 입력이 Input Sample과 같다면 [$1+3x$ , $2+4x$, $3$] 으로 생각하시면 됩니다.
즉, $ (1 + 3x) * (2 + 4x) * 3  = 6 + 30x + 36x^2 $
결과로 나옵니다. 이 결과에서 나오는 다항식의 계수를 공백으로 분리하여 오름차순으로 출력합니다. 즉, 6 30 36이 출력됩니다.


```
Tip 1) 다항식의 곱셈의 차수
최대 차수가 1차 * 1차 ⇒ 2차가 최대
최대 차수가 n차 * m차 ⇒ n+m차가 최대 
1차 함수의 곱 ⇒ 길이가 2인 리스트가 필요
2차 함수의 곱 ⇒ 길이가 3인 리스트가 필요
3차 함수 ⇒ 길이가 2 + 3 - 1 = 길이가 4인 리스트가 필요
```
```
Tip 2) 두 다항식을 곱한 결과의 i 번째 항의 계수?

ex) (2 + 3x) * (3 + 5x) 의 1차항의 계수 = 19 =
(앞의 항의 0차항 계수 2 * 뒤의 항 1차항의 계수 5 +
앞의 항의 1차항 계수 3 * 뒤의 항 0차항의 계수 3)

따라서 n + m = i 인 모든 n, m에 대하여 곱셈을 진행하자.
```
