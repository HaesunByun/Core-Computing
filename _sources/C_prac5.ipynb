### Problem: 가중치적용하여 성적구하기
😄 배열변수 scores에는 앞에서부터 번호, 국어, 영어, 수학, 체육, 컴기의 시험점수가 들어있습니다.

전체 평균점수가 가장 높은 학생 번호를 구하시오.

전체 평균점수가 가장 낮은 학생 번호를 구하시오.

국어, 영어, 수학 점수의 평균이 가장 높은 학생 번호를 구하시오.

국어, 영어, 수학 점수에 2배 가중치를 준다면 5과목의 평균 점수가 가장 높은 학생의 번호를 구하시오.



```
scores=np.array([[23231., 70., 85., 95., 100., 100.],
          [23232., 100., 84., 90., 90., 95.],   
          [23233., 95., 95., 80., 75., 75.],          
          [23234., 65., 75., 70., 90., 100.],
          [23235., 90., 80., 85., 95., 95.],          
          [23236., 90., 90., 95., 75., 80.],          
          [23237., 60., 90., 80., 40., 45.],          
          [23238., 85., 90., 75., 60., 90.],              
          [23239., 90., 85., 85., 90., 85.]])
```

---


<font color='blue'>`test case-1`</font>

Output Sample
```
전체 평균점수가 가장 높은 학생 번호:  23232

국,영,수 평균이 가장 높은 학생 번호:  23236
국,영,수 2배 가중치 평균이 가장 높은 학생 번호:  23232
```

````python
import numpy as np
scores=np.array([[23231., 70., 85., 95., 100., 100.],
          [23232., 100., 84., 90., 90., 95.],
          [23233., 95., 95., 80., 75., 75.],
          [23234., 65., 75., 70., 90., 100.],
          [23235., 90., 80., 85., 95., 95.],
          [23236., 90., 90., 95., 75., 80.],
          [23237., 60., 90., 80., 40., 45.],
          [23238., 85., 90., 75., 60., 90.],
          [23239., 90., 85., 85., 90., 85.]])
````
